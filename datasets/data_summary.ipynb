```python
import pandas as pd
import numpy as np
import os
```

```python
def get_data_summary(data_folder, group_by_key="", sep='\t',ignored_key="", ignored_con = ""):
    #iterate all files in data folder and print count by group_by_key
    files = []
    if data_folder:
        folder = data_folder
        for dir, _, filenames in os.walk(folder): 
            while len(filenames) > 0:
                f = filenames.pop()
                if os.path.isdir(os.path.join(dir, f)):
                    filenames.append(f)
                elif f.endswith(".tsv"):
                    files.append(os.path.join(dir, f))
   
    for file in files:
        
        data = pd.read_csv(file, sep = sep)
        if data.shape[0] == 0:
            continue
        print(file)
        print(data.shape)
        if ignored_key !="":
            num_ignored_tweets = data[data[ignored_key]==ignored_con].shape[0]
            print("#Ignored tweets ({}), #remaining tweets: {}".format(num_ignored_tweets, data.shape[0] - num_ignored_tweets))
            data = data[data[ignored_key]!= ignored_con]
        
        print(data.groupby(group_by_key).count()['tweet_id'])
        print(data.shape)
        print("...........................................")
```

```python
data_path = "/home/nguyen/tweet_interpret_sum/datasets/CrisisMMD_v2.0/annotations/"
get_data_summary(data_path, "text_human", ignored_key ='text_info', ignored_con = "")
```

```python
data_path = "/home/nguyen/tweet_interpret_sum/datasets/CrisisNLP_labeled_data_crowdflower"
get_data_summary(data_path, "label", ignored_key ='label', ignored_con = "not_related_or_irrelevant")
```
